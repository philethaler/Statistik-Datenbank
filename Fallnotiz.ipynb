Das Ziel dieses Notebooks ist es, die Unsicherheit von Punktschätzungen (Mittelwerte und Anteile) mittels Konfidenzintervallen zu quantifizieren. Dabei werden sowohl klassische (Wald) als auch simulationsbasierte Verfahren (Bootstrap) angewendet, um die Präzision der Schätzungen für Transaktionsbeträge und Betrugsraten zu beurteilen.

### Was wurde gemacht
 Es wurden 95%-Konfidenzintervalle für die durchschnittlichen Ausgaben nach Geschlecht (Mittelwerte im Bereich von ca. 67$ bis 71$) sowie für die allgemeine Betrugsrate berechnet. Hierbei kam das Wald-Konfidenzintervall zum Einsatz. Ergänzend wurde ein Bootstrap-Verfahren mit 2000 Resamples durchgeführt, um die Stichprobenverteilung unabhängig von Verteilungsannahmen zu visualisieren. Ergänzend wurde eine Konvergenzsimulation implementiert, die den Schätzprozess dynamisch abbildet. Dabei wurde für verschiedene Stichprobenumfänge (n) berechnet, wie schnell der Schätzfehler sinkt und ab welchem Punkt die Fehlerspanne (Margin of Error) die definierte Zielmarke von ±1.0 unterschreitet. Abschliessend wurde eine Stichprobenplanung durchgeführt, um das notwendige n für eine vordefinierte Fehlergrenze (Margin of Error) zu bestimmen.

### Erkenntnisse

- **Bootstrap-Verteilung:** Das Histogramm der Bootstrap-Mittelwerte zeigt trotz der geringeren Stichprobengrösse eine Annäherung an die Normalverteilung, was die Robustheit des zentralen Grenzwertsatzes unterstreicht.

- **Betrugsanteil:** Wir sehen dass die Schätzung sehr Präzise ist und das der Betrugsanteil sehr gering ist.

- **Durchschnitts Betrag Frau-Mann:** Die Analyse zeigt, dass sich die Konfidenzintervalle von Männern und Frauen überschneiden. Da die statistische Kraft (Power) bei über einer Million Beobachtungen extrem hoch ist, lässt diese Überschneidung den Schluss zu, dass kein praktisch relevanter Unterschied im Ausgabeverhalten der Geschlechter existiert.

- **Einfluss der Stichprobengrösse:** Die Untersuchung verdeutlicht den Zusammenhang zwischen Fallzahl und Präzision. Während bei kleineren Teilmengen (z. B. n=1000) noch Unsicherheit über die Differenz der Mittelwerte bestehen könnte, liefert die vorliegende Vollerhebung (Big Data) eine nahezu punktgenaue Bestätigung der Ähnlichkeit beider Gruppen

- **Dynamik der Präzision (Gesetz der großen Zahlen):** Die Simulation der Stichprobengröße verdeutlicht das Prinzip der Konsistenz: Mit steigendem n konvergiert der Stichprobenmittelwert sichtbar gegen den wahren Mittelwert der Grundgesamtheit. Besonders auffällig ist der „abnehmende Grenznutzen“ der Stichprobengröße – während die Präzision bei kleinen n (bis ca. 1.000) sprunghaft ansteigt, erfordert eine weitere Halbierung des Fehlers eine Vervierfachung des Stichprobenumfangs.


Ein Punktschätzer allein ist nicht aussagekräftig, doch in Kombination mit den berechneten Konfidenzintervallen und der Konvergenzsimulation wird die hohe Verlässlichkeit der Ergebnisse deutlich. Die Simulation zeigt eindrücklich, dass die Präzision ab einem gewissen Punkt (n) nur noch langsam steigt (abnehmender Grenznutzen). Dass sich die Intervalle trotz der extrem hohen statistischen Power bei über einer Million Beobachtungen überschneiden, ist ein starker Beleg dafür, dass das Geschlecht in diesem Datensatz kein differenzierendes Merkmal für die Transaktionshöhe darstellt. Ein Unterschied ist hier nicht nur statistisch nicht signifikant, sondern praktisch nicht existent.

## **Notebook:** Analyse_9.ipynb  
**Datum:** 01.12.2025  

Ziel dieser Analyse ist es, Gruppenvergleiche für eine numerische Variable durchzuführen und zu untersuchen, ob sich die Verteilungen der Transaktionsbeträge zwischen verschiedenen Gruppen systematisch unterscheiden. Dabei werden geeignete statistische Tests eingesetzt, um Unterschiede zwischen mehreren Gruppen objektiv zu identifizieren und inhaltlich zu interpretieren.

### **Transaktionskategorie (`category`) im Verhältnis zum Transaktionsbetrag (`amt`)**

**Analyse**
Die Transaktionsbeträge wurden zwischen allen Transaktionskategorien verglichen,
um zu untersuchen, ob sich das typische Ausgabenniveau systematisch zwischen
den Kategorien unterscheidet. Die Analyse basiert auf unabhängigen Gruppen mit
stark rechtsschief verteilten Beträgen.

**Statistik**
- Kruskal-Wallis-Test zur Überprüfung eines globalen Unterschieds zwischen
  allen Kategorien  
- Dunn-Post-hoc-Test mit FDR-Korrektur für paarweise Vergleiche  
- Effektgröße: Cliff’s Delta für die paarweisen Kategorienvergleiche

**Warum**
Der Kruskal-Wallis-Test ist geeignet, da keine Normalverteilung der
Transaktionsbeträge vorliegt und mehr als zwei unabhängige Gruppen verglichen
werden. Der Dunn-Post-hoc-Test identifiziert konkrete Unterschiede zwischen
einzelnen Kategorienpaaren, während die FDR-Korrektur das Problem multipler
Tests adressiert. Cliff’s Delta wird als Effektgröße verwendet, da es
nichtparametrisch ist und die Stärke der Unterschiede zwischen zwei Gruppen
robust quantifiziert.

**Erkenntnis**
Der globale Test ist hochsignifikant und weist auf systematische Unterschiede
zwischen den Transaktionskategorien hin. Die berechneten Werte von Cliff’s
Delta zeigen überwiegend kleine bis mittlere Effekte, was darauf hindeutet,
dass die Kategorie einen relevanten, aber nicht dominanten Einfluss auf die
Höhe der Transaktionsbeträge hat. Die meisten Kategorien unterscheiden sich
klar voneinander, während nur wenige ähnliche Ausgabenniveaus aufweisen.

### **Alter (`age`) im Verhältnis zum Transaktionsbetrag (`amt`)**

**Analyse**
Das Alter der Karteninhaber:innen wurde in mehrere Altersgruppen eingeteilt.
Anschließend wurden die Transaktionsbeträge zwischen diesen Altersgruppen
verglichen, um zu untersuchen, ob sich das Ausgabeverhalten systematisch
zwischen unterschiedlichen Alterssegmenten unterscheidet.

**Statistik**
- Kruskal-Wallis-Test zur Prüfung eines globalen Unterschieds zwischen den
  Altersgruppen  
- Dunn-Post-hoc-Test für paarweise Vergleiche zwischen Altersgruppen  
- Benjamini–Hochberg-FDR-Korrektur zur Kontrolle multipler Tests  
- Effektgröße: Cliff’s Delta für die paarweisen Altersgruppenvergleiche

**Warum**
Da die Transaktionsbeträge stark rechtsschief verteilt sind und mehrere
unabhängige Altersgruppen verglichen werden, wurde ein nichtparametrischer
Ansatz gewählt. Der Kruskal-Wallis-Test prüft zunächst, ob sich die Gruppen
insgesamt unterscheiden. Der anschließende Dunn-Test identifiziert konkrete
Unterschiede zwischen einzelnen Altersgruppen. Die Benjamini–Hochberg-
Korrektur reduziert das Risiko von Fehlinterpretationen durch multiples
Testen. Cliff’s Delta quantifiziert die Stärke der Unterschiede zwischen
jeweils zwei Altersgruppen ohne zusätzliche Verteilungsannahmen.

## **Notebook:** Analyse_11.ipynb

**Datum:** 08.12.2025

* Erweiterung der Features auf numerische (z-Standardisiert), kategorische (OneHotEncoded) sowie zusätzlich hinzugefügte (Haversine-Distanz und Burst-Rate)
* Nutzung eines Test-/Train-Splits sowie Cross-Validation, um die Modellperformance besser vorhersagen zu können
* Untersuchung der Multikollinearität sowie des Variance Inflation Factors & geeignete Massnahmen als Konsequenz (bündeln der Information)
* Wechsel zur Logistischen Regression für eine korrekte Wahrscheinlichkeitsmodellierung mit Korrektur der Klassen-Inbalance
* Gezielte Threshold-Wahl
* Erweiterung des Modells auf einen Panel-Ansatz zur Berücksichtigung der Veränderung zwischen aufeinanderfolgenden Transaktionen mittels First Differences

**Warum**
Die einfache lineare Regression stiess an ihre Grenzen. Deswegen erweiterten wir sie zuerst mit weiteren Features, um dem Modell mehr Möglichkeiten zu geben, eine Transaktion korrekt zu klassifizieren. Dazu nutzten wir standardmässige Preprocessing-Schritte (z-Standardisierung der numerischen Daten damit grössere absolute Zahlen das Modell nicht beeinflusse, OneHotEncoding für kategorische Features um diese korrekt numerisch zu enkodieren ohne perfekte Kollinearität) und entwickelten zwei zusätzliche Features, um die Distanz Käufer-Verkäufer zu modellieren (meist ist der Betrüger nicht am selben Ort wie der Verkäufer) sowie die Burst-Rate von Transaktionen über 24h abzubilden, da wir in den Hypothesentests gesehen haben, dass diese erhöht ist bei Betrugsfällen. Den Test-/Train-Split sowie die CV haben wir genutzt, um die Performance unseres Modells auf verschiedene (in unserem Fall 5) Folds d.h. Teile des Training-Datensatzes zu testen, um zufällig gute Ergebnisse ausschliessen zu können. Im Anschluss wurde die Performance auf den bisher komplett ungesehenen Test-Daten (30% des ganzen Datensatzes) getestet, um die zukünftige (Praxis-)-Performance des Modells realistisch sehe zu können. Die Multikollinearität wurde untersucht und behandelt, um die Koeffizienten im Modell stabiler zu halten, da das Modell so nicht Schwierigkeiten hat, die Effekte einem Feature zuzuordnen. Die logistische Regression als nächsten Schritt wurde gewählt, da die Vorhersage einer Betrugswahrscheinlichkeit immer zwischen 0 und 1 geschehen soll, was perfekt erfüllt wird von diesem Modell – ein lineares Regressionsmodell ist dort nicht beschränkt und kann auch tiefere/höhere Werte vorhersagen. Ebenso glichen wir im Code die extreme Klassen-Inbalance aus, um das Modell sensitiver auf Fraud zu machen (was unser Ziel ist). Zusätzlich wählten wir bewusst einen sehr tiefen Threshold zur Betrugs-klassifikation, um einen hohen Recall zu erreichen, was wichtiger in diesem Business Case ist für Nutzer des Modells, denn eine entgangene Fraud-Transaktion ist teurer als eine falsch-positiv markierte Nichtfraud-Transaktion. Zum Schluss wollten wir unbedingt noch ein Panel-Ansatz nutzen, und wählten die First Differences als perfekt geeignete Ergänzung, um Betrugsmuster, die oft Abweichungen vom gewohnten Verhalten darstellen, als Feature zu modellieren. 

**Ergebnis**
Die lineare Regression schloss schlechter ab als ein naives Modell, wahrscheinlich aufgrund der extremen Klassen-Imbalance und der hohen Anzahl an kategorialen Merkmalen.
Die logistische Regression war bereits viel besser und ca. 13x besser als der Zufall mit einer Average Precision von 7.8% bei einem gesetzten Recall von ca. 80%.
Durch die Ergänzung der First Difference Feature konnte das Modell nochmals signifikant verbessert werden, wir erreichten eine Average Precision von knapp 19% bei einer Erhöhung des Recalls auf 90% durch die Wahl eines höheren Thresholds. Interessanterweise zeigten sich dabei neben den kategorischen Features des Berufs bei den numerischen Features vor allem die Transaktionshöhe sowie eine höhere Differenz zum vorherigen Transaktionsbetrag als wichtige Prädiktoren, wobei interessanterweise eine grössere Differenz zum vorherigen Transaktionsbetrag eher auf eine geringere Wahrscheinlichkeit für Betrug hindeutet.
